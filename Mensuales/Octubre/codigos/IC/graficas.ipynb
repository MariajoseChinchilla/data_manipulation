```python
df = pd.read_csv("../../db/clientes_contagio.csv", encoding="cp1252", sep=";")
```

```python
df.columns
```

```python
x_labels = df['Mes'] + " " + df['Año'].astype(str)
x = np.arange(len(df))

# Crear la figura y los ejes
fig, ax1 = plt.subplots(figsize=(10,6))

# Graficar 'TICKET NETO' (equivalente a Cumplimiento en la imagen)
bars = ax1.bar(x, df['TICKET NETO'], color='#04aca4', alpha=0.7, label='Cumplimiento', zorder=1)
ax1.set_xlabel('Año')
ax1.set_ylabel('Ticket Neto', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_ylim([0, 140000])  # Ajusta este rango según tus datos
ax1.set_xticks(x)
ax1.set_xticklabels(x_labels, rotation=90)  # Etiquetas verticales

# Añadir etiquetas en cada barra (Ticket Neto) de forma intercalada y con formato de millares
for i, bar in enumerate(bars):
    yval = bar.get_height()
    # Ajustar la posición de la última etiqueta de las barras
    if i == len(bars) - 1:  # Última barra
        ax1.text(bar.get_x() + bar.get_width()/2, yval - 5000, f'{yval:,.2f}',  # Mover la etiqueta hacia abajo
                 ha='center', va='bottom', fontsize=8, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
    elif i % 2 == 0:  # Para las demás barras pares
        ax1.text(bar.get_x() + bar.get_width()/2, yval + 3000, f'{yval:,.2f}',  # Formato con comas para millares
                 ha='center', va='bottom', fontsize=8, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

# Añadir el segundo eje para '% TPP Cosecha Mensual' (equivalente a Metas en la imagen)
ax2 = ax1.twinx()
line, = ax2.plot(x, df['% TPP Cosecha Mensual'], color='#e2056b', marker='s', linewidth=2, label='Metas', zorder=2)
ax2.set_ylabel('% TPP Cosecha', color='black')
ax2.tick_params(axis='y', labelcolor='black')
ax2.set_ylim([15, 30])  # Ajusta el rango de acuerdo con tus datos

# Añadir etiquetas en cada punto de la curva (Metas) de forma intercalada y mover la última etiqueta
for i, value in enumerate(df['% TPP Cosecha Mensual']):
    if i % 2 == 0:  # Mostrar solo las etiquetas intercaladas
        if i == len(df['% TPP Cosecha Mensual']) - 1:  # Mover la última etiqueta de la curva hacia arriba
            ax2.text(x[i], value + 0.5, f'{value:.2f}', color='black', ha='center', va='bottom', fontsize=8,
                     bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
        else:
            ax2.text(x[i], value + 1, f'{value:.2f}', color='black', ha='center', va='bottom', fontsize=8,
                     bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

# Unir las leyendas en una sola línea y moverlas más a la izquierda y ligeramente hacia arriba para que no choquen
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left', bbox_to_anchor=(-0.25, 1.15), ncol=2, fancybox=True, shadow=True)

# Añadir el título similar al de la imagen
plt.title('Ticket neto y TPP Cosecha\ndesde enero 2023 a septiembre 2024', fontsize=16, weight='bold')

# Ajustar el layout para evitar superposición
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
