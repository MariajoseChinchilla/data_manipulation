```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../db/Macro variables financieras.xls", sheet_name="Pruebas estadísticas")
```

```python
df_temp = df[df["Inflación"] <= 0.06]
```

```python
plt.plot(figsize=(10,6))
plt.scatter(df_temp["Inflación"], df_temp["TC vigente"])
plt.title("TC vigente según inflación")
plt.show()
```

```python
plt.plot(figsize=(10,6))
plt.scatter(df_temp["Inflación"], df_temp["Préstamos consumo"])
plt.title("Préstamos de consumo según inflación")
plt.show()
```
