```python
def esperar(segundos=1):
    time.sleep(segundos)

# Función para iterar sobre los recuadros dinámicos
def click_dynamic_buttons(image_path: str="../db/boton.png", scroll_limit=10):
    scroll_count = 0

    while scroll_count < scroll_limit:  # Evitar loops infinitos
        # Intentar encontrar el botón en pantalla
        button_location = pyautogui.locateOnScreen(image_path, confidence=0.8)
        if button_location:
            # Si se encuentra el botón, dar clic en él
            pyautogui.click(button_location)
            esperar(2)  # Pausa para procesar

        # Hacer scroll hacia abajo y buscar nuevamente
        pyautogui.scroll(-500)  # Desplazarse hacia abajo
        scroll_count += 1
        esperar(2)
```
